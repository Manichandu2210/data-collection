### Asking Basic Questions (EDA)

1. how big is the data we are working with — `data.shape`
2. how is the data look — `data.head()` or `data.sample()`
3. what are the different data types we are working with — `df.info()`
4. do we have any missing values in the data — `df.isnull().sum()`
5. how does data look mathematically for statistical analysis — `df.describe()`
6. is there any duplicate columns in the data — `df.columns[df.columns.duplicated()]`
7. how is the correlation between the columns — `df.corr()`
