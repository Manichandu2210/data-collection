```python
import requests
import pandas as pd

base_url = 'https://pokeapi.co/api/v2/'
pokemon_data = []

# You can change the limit as needed — too high will slow it down
response = requests.get(f'{base_url}pokemon?limit=10000&offset=0')
if response.status_code == 200:
    print("Fetching Pokémon data...")
    results = response.json()["results"]

    for item in results:
        detail_url = item["url"]
        detail_response = requests.get(detail_url)
        if detail_response.status_code == 200:
            d = detail_response.json()

            # Extract core attributes
            name = d["name"]
            poke_id = d["id"]
            height = d["height"]
            weight = d["weight"]
            base_experience = d.get("base_experience", None)

            # Extract stats into a dict
            stats = {s["stat"]["name"]: s["base_stat"] for s in d["stats"]}

            # Extract types
            types = [t["type"]["name"] for t in d["types"]]
            type_1 = types[0] if len(types) > 0 else None
            type_2 = types[1] if len(types) > 1 else None

            # Extract abilities
            abilities = [a["ability"]["name"] for a in d["abilities"]]

            # Build record
            pokemon_data.append({
                "id": poke_id,
                "name": name,
                "height": height,
                "weight": weight,
                "base_experience": base_experience,
                "type_1": type_1,
                "type_2": type_2,
                "hp": stats.get("hp"),
                "attack": stats.get("attack"),
                "defense": stats.get("defense"),
                "special_attack": stats.get("special-attack"),
                "special_defense": stats.get("special-defense"),
                "speed": stats.get("speed"),
                "abilities": abilities
            })

df = pd.DataFrame(pokemon_data)
print(df.head())

# Optional: save to CSV
df.to_csv("pokemon_full_data.csv", index=False)
print("Saved to pokemon_full_data.csv")
```
