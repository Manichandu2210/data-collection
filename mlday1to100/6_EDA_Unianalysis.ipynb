```python
df['Survived'].unique()
```

### Working with the categorical data

```python
sns.countplot(x = df['Survived'],data = df,palette= 'deep',hue='Sex')
```

```python
df['Sex'].value_counts().plot(kind='pie',legend=True,autopct='%.2f')
```

```python
sns.histplot(df['Age'],kde=True,bins = 5)
```

```python
sns.displot(df['Age'],bins = 5,kde = True)  
```
