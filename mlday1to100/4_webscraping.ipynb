```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

base_url = "https://books.toscrape.com/catalogue/category/books/"
main_url = "https://books.toscrape.com/catalogue/category/books_1/index.html"

response = requests.get(main_url).text
soup = BeautifulSoup(response, 'lxml')

# Step 1: Get all category links
categories = soup.find('ul', class_='nav-list').find_all('a')
data = []

for category in categories:
    genre = category.text.strip()
    link = "https://books.toscrape.com/" + category['href']
    genre_page = requests.get(link).text
    genre_soup = BeautifulSoup(genre_page, 'lxml')
    
    # Step 2: Find all books in that category
    books = genre_soup.find_all('article', class_='product_pod')
    for book in books:
        title = book.h3.a['title']
        data.append({"Genre": genre, "Title": title})

# Step 3: Store it
df = pd.DataFrame(data)
print(df)

df.to_csv('books data')
```
