#### When we are working on the single variable(column) is called Univariate analysis

```python
import pandas as pd 
import numpy as np
import seaborn as sns

df = pd.read_csv(r'csv/titanic.csv')
```

```python
df['Survived'].unique()
```

```python
df['Survived'].value_counts()
```

### Working with the categorical data

```python
sns.countplot(x = df['Survived'],data = df,palette= 'deep',hue='Sex')
```
